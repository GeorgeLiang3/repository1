```python
from scipy.stats import multivariate_normal
x, y = np.mgrid[-4:4:.01, -6:4:.01]
pos = np.empty(x.shape + (2,)) 
pos[:, :, 0] = x; pos[:, :, 1] = x**2+y
rv = multivariate_normal([0, 0], [[1, 0], [0, 0.3]])
plt.contour(x, y, rv.pdf(pos))
plt.xlabel("x1")
plt.ylabel("x2")
```

```python
#Randaom walk Matropolis

initial = [0,0]

rv.pdf(initial)
```

```python
#Random Walk MH MCMC
@tf.function
def draw_proposal(m_current):
    
    Xi = tfd.MultivariateNormalFullCovariance(
            loc = m_current,
            covariance_matrix= [[0.1,0],[0,0.1]])
    Xi_s = tfd.Sample(Xi)
    m_proposed = Xi_s.sample()
    
    return m_proposed
    

def accept_or_not(m_current,m_proposed,result):
    if not isinstance(m_current,list):m_current = m_current.tolist()
    if not isinstance(m_proposed,list):m_proposed = m_proposed.tolist()
    
    p_accept = rv.pdf([m_proposed[0],m_proposed[0]+m_proposed[1]**2])/rv.pdf([m_current[0],m_current[0]+m_current[1]**2])
    accept = np.random.rand() < p_accept
    print(accept)
    if accept:
        m_current = m_proposed
        result.append([m_proposed,1])

    else:
        result.append([m_proposed,0])

        
```

```python
rv.pdf(m_proposed)
```

```python
rv.pdf([m_proposed[0],m_proposed[0]+m_proposed[1]**2])
```

```python
def chain(step = 10,resultlist=[], Session = sess):

    initial = [-2, -4]
    m_current = initial
    for s in range(step):
        m_proposed = draw_proposal(m_current)
        m_proposed = Session.run(m_proposed)
        accept_or_not(m_current,m_proposed,result)

        
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    result = []
    chain(resultlist = result, Session = sess)
```
