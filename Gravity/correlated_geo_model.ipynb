## Simple correlated geological model

```python
import numpy as np
import scipy.interpolate as sci
import matplotlib.pyplot as plt
```

```python
sci.interp1d?
```

```python
xvals = np.arange(5.1)
yvals = np.array([1,4,3,1,0,2])
```

```python
plt.plot(xvals, yvals, 'o:')
```

```python
f = sci.interp1d(xvals, yvals, kind='cubic')
```

```python
xx = np.arange(0,5,0.1)
```

```python
plt.plot(xvals, yvals, 'o:')
plt.plot(xx, f(xx), 'r--')
```

```python
x_interp_pts = np.linspace(0,5,10)
```

```python
plt.plot(xvals, yvals, 'o:')
plt.plot(xx, f(xx), 'r--')
plt.plot(x_interp_pts, f(x_interp_pts), 'gs')
for xxx in x_interp_pts:
    plt.axvline(xxx, c='k', linestyle=':')
```
